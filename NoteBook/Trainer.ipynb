```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier , RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

from xgboost import XGBClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import accuracy_score,f1_score,classification_report,recall_score
```

```python
df=pd.read_csv(r'D:\Data_Science\7-Machine_Learning\projects\END-TO-END_projectsML\Customer Churn Prediction_END_TO_END_ML\NoteBook\Churn.csv')
```

```python
df.head()
```

```python
x=df.drop('Churn',axis=1)
y=df['Churn']

print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.1)
```

```python
x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)


x_train.fillna(0, inplace=True)
x_test.fillna(0, inplace=True)
```

```python
def evaluate_model(y_test,predict):
    f1=f1_score(y_test,predict)
    report=classification_report(y_test,predict)
    accuracy=accuracy_score(y_test,predict)
    recall=recall_score(y_test,predict)
    
    return f1 , report , accuracy, recall
```

```python
models={
    "LogisticRegression":LogisticRegression(),
    "Knn":KNeighborsClassifier(),
    "Tree":DecisionTreeClassifier(),
    "Randomforest":RandomForestClassifier(),
    "AdaBoost":AdaBoostClassifier(),
    "XGBoost":XGBClassifier(),
    "CatBoost":CatBoostClassifier()
}

model_list=[]
evaluate_list=[]

for name,obj_model in models.items():
    obj_model.fit(x_train,y_train)

    predict_train=obj_model.predict(x_train)
    predict_test=obj_model.predict(x_test)

    f1_train,report_train,accuracy_train,recall_train=evaluate_model(y_train,predict_train)
    f1_test,report_test,accuracy_test,recall_test=evaluate_model(y_test,predict_test)
    
    print(name)

    model_list.append(name)

    print("f1_train",f1_train)
    print("report_train",report_train)
    print("accuracy_train",accuracy_train)
    print("recall_train",recall_train)
    print('-'*14)
    print("f1_test",f1_test)
    print("report_test",report_test)
    print("accuracy_test",accuracy_test)
    print("recall_test",recall_test)  

    print("="*25)
    print('\n')


    evaluate_list.append((recall_test, f1_test))

```

```python
result_df=pd.DataFrame({
    "model_list":model_list,
    "evaluate_list":evaluate_list
})
```

```python
best_name_model=result_df.iloc[0]['model_list']
best_accuracy_model=result_df.iloc[0]['evaluate_list']
```

```python
print("The Best Model Is Name : ",best_name_model)
print("And Accuracy is equal = ",best_accuracy_model)
```

```python
model_linear=LogisticRegression()
model_linear.fit(x_train,y_train)
predict=model_linear.predict(x_test)
f1_test,report_test,accuracy_test,recall_test=evaluate_model(y_test,predict)

print("f1_test",f1_test)
print("report_test",report_test)
print("accuracy_test",accuracy_test)
print("recall_test",recall_test)  
```
